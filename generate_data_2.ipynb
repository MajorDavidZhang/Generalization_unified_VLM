```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import patches
from PIL import Image
import random

# 设置字体
from matplotlib import font_manager
font_path = '/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf'

# 随机生成参数
def random_time():
    hours = random.randint(1, 12)
    minutes = random.randint(0, 59)
    seconds = random.randint(0, 59)
    return hours, minutes, seconds

def random_color():
    return np.random.rand(3,)

def random_material():
    materials = ['metal', 'wood', 'plastic']
    return random.choice(materials)

def random_text_color():
    return np.random.rand(3,)

def random_border_color():
    return np.random.rand(3,)

# 绘制时钟
def draw_clock():
    fig, ax = plt.subplots(figsize=(2.24, 2.24), dpi=100)

    # 随机选择参数
    hours, minutes, seconds = random_time()
    hour_color = random_color()
    minute_color = random_color()
    second_color = random_color()
    hour_material = random_material()
    minute_material = random_material()
    second_material = random_material()
    text_color = random_text_color()
    border_color = random_border_color()

    # 创建表盘
    ax.add_patch(plt.Circle((0.5, 0.5), 0.45, color='white', ec='black', lw=2))

    # 绘制时刻数字
    for i in range(1, 13):
        angle = np.deg2rad(360 * (i / 12))
        x = 0.5 + 0.38 * np.cos(angle)
        y = 0.5 + 0.38 * np.sin(angle)
        ax.text(x, y, str(i), color=text_color, ha='center', va='center', fontproperties=font_manager.FontProperties(fname=font_path, size=10))

    # 绘制表圈
    ax.add_patch(plt.Circle((0.5, 0.5), 0.45, color=border_color, lw=6, fill=False))

    # 绘制指针
    # 时针
    hour_angle = np.deg2rad(360 * (hours % 12 + minutes / 60) / 12)
    ax.add_patch(patches.FancyArrowPatch((0.5, 0.5), (0.5 + 0.2 * np.cos(hour_angle), 0.5 + 0.2 * np.sin(hour_angle)), 
                                        mutation_scale=15, color=hour_color))

    # 分针
    minute_angle = np.deg2rad(360 * (minutes + seconds / 60) / 60)
    ax.add_patch(patches.FancyArrowPatch((0.5, 0.5), (0.5 + 0.3 * np.cos(minute_angle), 0.5 + 0.3 * np.sin(minute_angle)), 
                                        mutation_scale=10, color=minute_color))

    # 秒针
    second_angle = np.deg2rad(360 * seconds / 60)
    ax.add_patch(patches.FancyArrowPatch((0.5, 0.5), (0.5 + 0.35 * np.cos(second_angle), 0.5 + 0.35 * np.sin(second_angle)),
                                        mutation_scale=5, color=second_color))

    # 去除坐标轴
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal', 'box')
    ax.axis('off')

    # 保存图片
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0)
    plt.margins(0, 0)
    plt.savefig("clock_image.png", dpi=100)
    plt.close()

# 生成并显示图像
draw_clock()

# 显示生成的图片
img = Image.open('clock_image.png')
img.show()
```
